```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('zomato.csv',encoding='latin-1')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
[features for features in df.columns if df[features].isnull().sum()>0]
```

```python
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
## Merging the 2 datasets

final_df=pd.merge(df,df_country,on='Country Code', how='left')
final_df.head()
```

```python
country_names=final_df.Country.value_counts().index
country_names
```

```python
country_values=final_df.Country.value_counts().values
country_values
```

```python
plt.pie(country_values[:3],labels=country_names[:3],autopct='%1.2f%%')
```

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Rating count', data = ratings)
```

```python
sns.barplot(x='Aggregate rating',y='Rating count',hue = 'Rating color', data = ratings,palette=['white','red','orange','yellow','green','grey'])
                                  
```

```python
sns.countplot(x='Rating color',data=ratings,palette=['white','red','orange','yellow','green','grey'])
```
