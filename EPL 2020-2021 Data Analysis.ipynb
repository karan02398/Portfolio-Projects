```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
epl_df=pd.read_csv('epl.csv')
epl_df.head()
```

```python
epl_df.info()
```

```python
epl_df.describe()
```

```python
## Lets check for Null values

epl_df.isnull().sum()
```

###### As we can clearly see that there are no NaN values is the dataset.

```python
## Wecreated 2 new columns for Minutes played per match and goals scored per match

epl_df['MinsPerMatch']=(epl_df['Mins']/epl_df['Matches']).astype(int)
epl_df['GoalsPerMatch']=(epl_df['Goals']/epl_df['Matches']).astype(float)
epl_df.head()
```

```python
## Total goals scored in the season

Total_goals=epl_df['Goals'].sum()
Total_goals
```

```python
## penalty goals

penalty_goals=epl_df['Penalty_Goals'].sum()
penalty_goals
```

```python
## penalty attempts

penalty_attempted=epl_df['Penalty_Attempted'].sum()
penalty_attempted
```

```python
## pie-chart for penalties scored VS missed

plt.figure(figsize=(13,6))
pl_not_scored=epl_df['Penalty_Attempted'].sum()-epl_df['Penalty_Goals'].sum()
data=[pl_not_scored,penalty_goals]
labels=['Penalties missed','Penalties scored']
color=sns.color_palette('Set2')
plt.pie(data,labels=labels,colors=color,autopct= '%.0f%%')
plt.show()
```

## 82% of the penalties were scores while 18% were missed

```python
nationality=epl_df.groupby('Nationality').size().sort_values(ascending=False)
nationality.head(10).plot(kind='bar',figsize=(12,6),color=sns.color_palette('magma'))
```

```python
epl_df['Club'].value_counts().head(10).plot(kind='bar',color=sns.color_palette('viridis'))
```

```python
epl_df['Club'].value_counts(ascending=True).nsmallest(10).plot(kind='bar',color=sns.color_palette('viridis'))
```

```python
## Players based on age groups

under20=epl_df[epl_df['Age']<=20]
age20_25=epl_df[(epl_df['Age']>20) & (epl_df['Age']<=25)]
age25_30=epl_df[(epl_df['Age']>25) & (epl_df['Age']<=30)]
above30=epl_df[epl_df['Age']>30]
```

```python
x=np.array([under20['Name'].count(),age20_25['Name'].count(),age25_30['Name'].count(),above30['Name'].count()])
mylabels=['<20','>20 &<=25','>25 & <=30','>30']
plt.title('Total players with age Groups',fontsize=20)
plt.pie(x,labels=mylabels,autopct='%.1f%%')
plt.show()
```

```python
## Avg age of players in each club

plt.figure(figsize=(12,6))
sns.boxplot(x='Club',y='Age',data=epl_df)
plt.xticks(rotation=90)
```

```python
## Top 10 assists by players

top10_assists=epl_df[['Name','Club','Assists','Matches']].nlargest(n=10,columns='Assists')
top10_assists
```

```python
goals_club=pd.DataFrame(epl_df.groupby('Club',as_index=False)['Goals'].sum())
sns.set_theme(style='whitegrid',color_codes=True)
ax=sns.barplot(x='Club',y='Goals',data=goals_club.sort_values(by='Goals'),palette='rocket')
ax.set_xlabel('Club',fontsize=20)
ax.set_ylabel('Goals',fontsize=20)
plt.xticks(rotation=675)
plt.rcParams['figure.figsize']=(20,8)
plt.title('Plot of clubs VS Total Goals',fontsize=20)
```

```python
top10_goals=epl_df[['Name','Club','Goals','Matches']].nlargest(n=10,columns='Goals')
top10_goals
```

```python
top10_goals_permatch=epl_df[['Name','GoalsPerMatch','Goals','Matches']].nlargest(n=10,columns='GoalsPerMatch')
top10_goals_permatch
```

```python
## Pie-chart-Goals with assist and without assist

plt.figure(figsize=(14,7))
assists=epl_df['Assists'].sum()
data= [Total_goals-assists,assists]
labels=['Goals without assists','Goals with assists']
color=sns.color_palette('Set1')
plt.pie(data,labels=labels,colors=color,autopct='%.0f%%')
plt.show()
```

```python
## Top 10 players with most yelow card

epl_yellow=epl_df.sort_values(by='Yellow_Cards',ascending=False)[:10]
plt.figure(figsize=(20,6))
plt.title('Players with most Yellow card')
c=sns.barplot(x=epl_yellow['Name'],y=epl_yellow['Yellow_Cards'],label='Players',color='yellow')
plt.ylabel('Number of yellow cards')
c.set_xticklabels(c.get_xticklabels(),rotation=45)
c
```
